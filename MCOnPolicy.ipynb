```python
#@title Importamos librerias
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import gymnasium as gym
import Box2D
```

```python
#@title Importamos el lago helado
name = 'Taxi-v3'

# Creamos el entorno Taxi-v3
env = gym.make(name, render_mode="ansi")
```

```python
class MonteCarloAgent:
    def __init__(self, env, epsilon=0.1, alpha=0.1, gamma=0.99):
        self.env = env
        self.epsilon = epsilon  # Probabilidad de exploración
        self.alpha = alpha  # Tasa de aprendizaje
        self.gamma = gamma  # Factor de descuento
        self.Q = {}  # Tabla de valores Q (estado, acción)
        self.policy = {}  # Política exploratoria
    
    def init(self):
        # Inicializar la tabla Q y la política para todos los estados posibles
        for state in range(self.env.observation_space.n):
            for action in range(self.env.action_space.n):
                self.Q[(state, action)] = 0.0
            self.policy[state] = self.epsilon_greedy(state)
    
    def epsilon_greedy(self, state):
        # Política epsilon-greedy: selecciona acción aleatoria con epsilon, else la mejor acción
        if np.random.rand() < self.epsilon:
            return np.random.choice(self.env.action_space.n)  # Acción aleatoria
        else:
            return np.argmax([self.Q[(state, a)] for a in range(self.env.action_space.n)])

    def get_action(self, state):
        # Obtener la acción según la política
        return self.policy[state]
    
    def update(self, episode):
        # Actualizar los valores de Q usando las recompensas del episodio
        G = 0  # Recompensa total
        for state, action, reward in reversed(episode):
            G = self.gamma * G + reward
            self.Q[(state, action)] += self.alpha * (G - self.Q[(state, action)])
            
    def get_Q(self):
        Q_matrix = np.zeros((self.env.observation_space.n, self.env.action_space.n))
        for (state, action), value in self.Q.items():
            Q_matrix[state, action] = value
        return Q_matrix
```

```python
agent = MonteCarloAgent(env)
agent.init()

# Ejecutar episodios de entrenamiento
rewards_per_episode = []  # Almacena las recompensas por episodio
episode_sizes = []  # Almacena el tamaño de cada episodio

# Ejecutar episodios de entrenamiento
for episode in tqdm(range(1000)):
    state, _ = env.reset()  # Ajuste aquí para que el estado se obtenga correctamente
    done = False
    episode_data = []
    total_reward = 0  # Acumulador de recompensas para el episodio
    episode_length = 0  # Contador de pasos por episodio
    
    while not done:
        action = agent.get_action(state)
        next_state, reward, done, _, _ = env.step(action)
        
        # Guardar el episodio (estado, acción, recompensa)
        episode_data.append((state, action, reward))
        total_reward += reward
        episode_length += 1
        
        state = next_state
    
    agent.update(episode_data)  # Actualizar el agente después de cada episodio
    
    # Almacenar recompensas y tamaño del episodio
    rewards_per_episode.append(total_reward)
    episode_sizes.append(episode_length)
```

```python
# Al final del entrenamiento, mostrar los resultados
print("Tabla Q:")
for state in range(env.observation_space.n):
    for action in range(env.action_space.n):
        print(f"Q({state}, {action}) = {agent.Q[(state, action)]}")

# Mostrar los resultados finales (recompensas y tamaño de episodios)
print("\nRecompensas por episodio:")
print(rewards_per_episode)

print("\nTamaño de episodios:")
print(episode_sizes)
```

```python
# Política Greedy a partir de los valones Q. Se usa para mostrar la solución.
def pi_star_from_Q(env, Q):
    done = False
    pi_star = np.zeros([env.observation_space.n, env.action_space.n])
    state, info = env.reset() # start in top-left, = 0
    actions = ""
    while not done:
        action = np.argmax(Q[state, :])
        actions += f"{action}, "
        pi_star[state,action] = action
        state, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
    return pi_star, actions
```

```python
# @title Política final
pi, actions = pi_star_from_Q(env, Q)

print("Política óptima obtenida\n", pi, f"\n Acciones {actions} \n Para el siguiente grid\n", env.render())
print()
```
